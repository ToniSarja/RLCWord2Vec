```python
import requests
from bs4 import BeautifulSoup
all_data = []
# Base URL (without the page number at the end)
base_url = "http://web-corpora.net/RLC/search/?exact_word=&wordform%5B%5D=&lex%5B%5D=%28S%7CA%7CNUM%7CANUM%7CV%7CADV%7CPARENTH%7CPRAEDIC%7CADVPRO%7CSPRO%7CAPRO%7CPR%7CCONJ%7CPART%7CINTJ%29&grammar%5B%5D=&errors%5B%5D=&from%5B%5D=1&to%5B%5D=1&wordform%5B%5D=&lex%5B%5D=&grammar%5B%5D=&errors%5B%5D=&lex_search=Search&date1=&date2=&gender=any&mode=any&background=FL&format=full&per_page=300&expand=%2B-1&page="
def scrape_page(page_number):
# Function to scrape a single page
    url = base_url + str(page_number)
    print(f"Scraping page {page_number}: {url}")
    
    # Send a request to the page
    
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        plain = soup.get_text()
        all_data.append(plain)
    else:
        print("Failed to retrieve the web page.")



for page in range(201, 470):  # Replace 11 with the total number of pages you want to scrape
    page_data = scrape_page(page)
    
# Example: Save the data to a CSV file or print the output
import csv

with open('corpus_data.txt', 'a', newline='\n', encoding='utf-8') as f:
    f.write(str(all_data))

print("Scraping complete!")
```
